## Models Training

- I tested 3 different archetectures with pytorch.

    1. ___Net3Conv___ - simple custom architecture with 3 convolution blocks and ~30k parameters.
    2. ___Net9Conv___ - deeper custom architecture with 9 convolution blocks and more parameters ~150k.
    3. ___Efficientnet-b0___ - with ~4M params
        
        [paper](https://arxiv.org/abs/1905.11946)
        
        [pytorch implementation](https://github.com/lukemelas/EfficientNet-PyTorch)


- Detailed architectures presented in ___section 7 [Networks  Architecture]___.

- Results presented in next ____section [Models Results]____

- I also tested a limited number of different training setups: learning rate/ epochs/ batch size/ optimizer algorithms.

- All networks tested with and without augmentation. For EfficientNet I also tested different levels of dropout.

### EfficientNet-B0 - TEST ACC=92.7%
This is much "fatter" network with ~4M parameters.

Without augmentations it got a similar overfitting as simpler networks (gap is 6%-7% in train/validation accuracy). 
I expected for more severe over-fitting (bigger net). The reason it is not occur is a dropout used in the efficient-Net (for B0 dropout_p=20%).


<img src="resources/plot_EfficientNet_Feb24_2254_aug_fasle.png" alt="Drawing" style="width: 500px;"/>

### EfficientNet-B0 - with augmentations and dropout TEST ACC=93.5%

The best one I I succeeded to produce with efficientNet-b0, testing a couple configurations with different lavels of augmentationsa and dropout.

Bigger model require a more rigorous training procedure. 

I belive that with better optimizations of learning rate, augmentations level, dropout and more epochs of training (4M params definetelly need more training than 150k like in Net9Conv), EfficientNet-B0 can outperform a simpler Net9Conv.

## Demo app (Bonus task)

___Testing Data___: 

color e-commerce clothing images from kaggle datasets [link](https://www.kaggle.com/paramaggarwal/fashion-product-images-dataset).
The apparels is in the center of the image.

___Data preprocessing___: 

Since the testing images is very different from Fashion-MNIST data, the preprocessing step is very important. Preprocessing steps:

1) Resize shortest side to 28 pixels 

2) Center crop 

3) Convert to gray scale with 1 channel 

4) Normilize to [0-1]. 

5) Last important stage is converting white background to black as in Fashion-MNIST. I simply did `img = 1.0 - img`


___Demo running instructions___:

 - Should be self-explanatory. 

 - All needed params is configurable in `demo_app.py/CONFIG` dict.

 - Attached 16 testing images and pre-trained net.

__Personal impression__:

Better than I expected. 

It detect correct the most of the images, of course if it is one of the 10 trained classes. (Testing images also contains clocks, jewelry and some other objects not presented in training data). 

False alarms mainly occur between similar classes, e.g. confusion between "Sneaker" and "Ankle boot" or "T-shirt/top" and "Shirt".

Possible next steps presented an section [5].

## Conclusions and Next Steps

*Main conclusion is 4 Millions parameterss is an overkill for Fashion-MNIST 28x28 images! Simple data require simple tool to deal with.*


___Possible next steps in order to improve results___:

1) I tested the less complicated EfficientNet - b0. But it make sense to build a custom EfficientNet with even less blocks and parameters for that kind of data.


2) Error analysis. Understand the cases and reasons for network mistakes.

3) Understand the difference between train and test set. Even simple model suffer  from over-fitting, possible reason is some kind of dissimilarity between train and test data.

5) *demo app* - Main reason for some false alarms is a difference between F-MNIST data and data for demo. In demo data the closes frequently appears on a human model, in difference from training F-MNIST data. Try to remove pixels with color of human skin - pain it with background color.

6) *demo app* - Implement a sliding window with different scales. In such way we can work with any images. Not only those where apparel in center of an image.
